```python
data['max_power'].str.contains("bhp").size-data['max_power'].size
data.loc[:,'max_power'] = data['max_power'].str[:-4].astype(float)
data.loc[:,'engine']=data['engine'].str[:-3].astype(int)
```

```python
data.loc[:,'Company']=data['name'].str.split(" ").str[0]
data.loc[:,'CarName']=data['name'].str.split(n=1).str[1].str.split(" ").str[0]
data
```

```python
data.loc[:,'Age']=2025-data.loc[:,'year']
data
```

```python
data['owner'].unique()
owners={'Test Drive Car':0,'First Owner':1,'Second Owner':2, 'Third Owner':3,
       'Fourth & Above Owner':4}
owners
data['owner'].map(owners)
```

```python
data.loc[:,'owner']=data['owner'].map(owners)
data
```

```python
data['seller_type'].unique()
```

```python
seller=pd.get_dummies(data['seller_type']).astype(int)
data=pd.concat([data,seller],axis='columns')
data
```

```python
data['fuelunit']=np.where(data['mileage'].str.contains('kmpl'),'kmpl','km/kg')
data['mileage']=np.where(data['fuelunit']=='kmpl',data['mileage'].str[:-5],data['mileage'].str[:-6])
```

```python
fuel_prices = {'Petrol': 100,'Diesel': 90,'CNG': 75,'LPG': 65}

data['mileage']=data['mileage'].astype(float)
#data.info()

data.loc[data['mileage']==0].CarName
m=data.groupby('CarName')['mileage'].mean()
```

```python
data['mileage']=data['mileage'].where(data['mileage']>0.0,data.groupby('CarName')['mileage'].transform('mean'))
```

```python
data.drop(axis=0,index=7337,inplace=True)
```

```python
data.drop('torque',axis=1,inplace=True)
```

```python
data['cost/km']=data.apply(lambda x: (fuel_prices[x['fuel']]/x['mileage']),axis=1)
```

```python
databackup=data.copy()
data
```

```python
trans=pd.get_dummies(data['transmission']).astype(int)
data=pd.concat([data,trans],axis='columns')
```

```python
f=pd.get_dummies(data['fuel']).astype(int)
data=pd.concat([data,f],axis='columns')
```

```python
databackup=data.copy()
```

```python
data.drop(['name','year','fuel','seller_type','transmission','fuelunit'],axis=1,inplace=True)
```

```python
data=data.reindex(labels=['selling_price', 'Company', 'CarName', 'km_driven', 'mileage','cost/km','engine','max_power','owner',
       'seats', 'Age', 'Dealer', 'Individual',
       'Trustmark Dealer','Automatic', 'Manual', 'CNG', 'Diesel',
       'LPG', 'Petrol'],axis=1)
data.drop('mileage',axis=1,inplace=True)
```

```python
data.to_csv('Pre_processed_data.csv')
```
